
**HipÃ³tesis**

(X) masa corporal, peso, medida de la cintura, edad

(Y) riesgo de hipertensiÃ³n

 Las personas con mayor peso, mayor masa corporal y una circunferencia de la cintura mÃ¡s grande tienen mÃ¡s riesgo de sufrir hipertensiÃ³n. Por eso
  mismo se da a entender wue  las personas que hacen mÃ¡s actividad fÃ­sica tienen menos riesgo de tener hipertensiÃ³n.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['masa_corporal', 'medida_cintura', 'peso', 'edad']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```

El coeficiente de determinacion nos dice que Y esta sujeta a X con un 35% de variaciÃ³n.

```python
b0, b1, b2, b3, b4= modelo.params

Fun = lambda X1, X2, X3, X4 : b0 + b1 * X1 + b2 * X2 + b3 * X3 +b4 +X4

Fun(0, 0.5, 1, 2)
```

# **Supuestos**

```python
from scipy.stats import shapiro
_, valor_p_shapiro = shapiro(modelo.resid)
print(f'Valor-p de Shapiro-Wilk: {valor_p_shapiro: 0.4f}\n')
```

Nos dice que los residuos del modelo no siguen una distribuciÃ³n normal, y por eso no se cumple la normalidad en la regresiÃ³n lineal.

```python
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(modelo.resid, X_train_constante)
print(f'Valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}')
```

Se rechaza la Ho, ya que es menor a 0.05.

Lo que nos dice que existe heterocedasticidad.

En este modelo anova se puede ver que la masa corporal, la edad y el colesterol estÃ¡n muy relacionados con la hipertensiÃ³n, pero el sexo no muestra una influencia alta en este anÃ¡lisis.



# **ConclusiÃ³n**

El anÃ¡lisis nos dice que la masa corporal, la edad y el colesterol total son factores significativos en el riesgo de hipertensiÃ³n, mientras que el sexo no muestra influencia.
En el coeficiente de determinaciÃ³n (RÂ² = 0.35),  solo explica el 35% de la variabilidad.
Como conclusiÃ³n, puedo decir que fueron pocas las variables que no se relacionan con el riesgo de hipertension, el mapa de calor, asi mismo nos muestra que lo que mas se reacionan son el peso, la medida de la cintura, la masa corporal, la edad, lo que se puede interpretar que las personas que sufren de hipertencion son las personas que tiene malos habitos con respecto a su peso.
