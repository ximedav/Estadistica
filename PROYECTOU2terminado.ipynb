# **Marco Te√≥rico: Regresi√≥n Lineal M√∫ltiple y Correlaci√≥n**

## Regresi√≥n Lineal M√∫ltiple

La regresi√≥n m√∫ltiple comprende de tres o m√°s variables. Existe una sola variable dependiente, pero hay dos o m√°s de tipos independiente (explicativo).

La **Regresi√≥n Lineal M√∫ltiple (RLM)** es una t√©cnica estad√≠stica que extiende el modelo de regresi√≥n lineal simple al considerar **m√°s de una variable independiente** para predecir una variable dependiente. El objetivo es encontrar una relaci√≥n lineal entre la variable dependiente \( Y \) y un conjunto de variables independientes \( X_1, X_2, ..., X_n \).

El modelo general se expresa como:

\[
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n + \error
\]

Formula : ùëå=ùõΩ0 + ùõΩ1ùëã1 +ùõΩ2ùëã2 + ‚Ä¶ + ùõΩùëõùëãùëõ + ùúñ


Donde:
- \( Y \) es la variable dependiente (respuesta),
- \( \beta_0 \) es la ordenada al origen (intercepto),
- \( \beta_1, \beta_2, ..., \beta_n \) son los coeficientes de regresi√≥n (pendientes),
- \( X_1, X_2, ..., X_n \) son las variables independientes (predictoras),
- \( \error \) es el t√©rmino de error.

### Supuestos del Modelo de RLM

1. Linealidad entre las variables independientes y la dependiente.
2. Independencia de los errores.
3. Homocedasticidad (varianza constante del error).
4. Normalidad de los errores.
5. No multicolinealidad entre las variables independientes.

### Evaluaci√≥n del Modelo

- **\( R^2 \) (Coeficiente de determinaci√≥n):** Indica qu√© proporci√≥n de la variaci√≥n de \( Y \) es explicada por las variables independientes.
- **Prueba F:** Eval√∫a si el modelo en su conjunto es significativo.
- **p-values:** Eval√∫an la significancia de cada coeficiente.

## Correlaci√≥n

La **correlaci√≥n** mide la fuerza y direcci√≥n de la relaci√≥n lineal entre dos variables cuantitativas. El coeficiente de correlaci√≥n m√°s com√∫n es el **coeficiente de correlaci√≥n de Pearson (r)**, que var√≠a entre -1 y 1:

- \( r = 1 \): Correlaci√≥n positiva perfecta.
- \( r = -1 \): Correlaci√≥n negativa perfecta.
- \( r = 0 \): No hay correlaci√≥n lineal.

### Interpretaci√≥n del Valor de r

| Valor de r | Interpretaci√≥n              |
|------------|-----------------------------|
| 0.9 ‚Äì 1.0  | Correlaci√≥n muy fuerte      |
| 0.7 ‚Äì 0.9  | Correlaci√≥n fuerte          |
| 0.5 ‚Äì 0.7  | Correlaci√≥n moderada        |
| 0.3 ‚Äì 0.5  | Correlaci√≥n d√©bil           |
| 0.0 ‚Äì 0.3  | Correlaci√≥n muy d√©bil o nula|

### Aplicaciones

La RLM y la correlaci√≥n se utilizan ampliamente en diversas √°reas como:
- Econom√≠a (para predecir consumo, inversi√≥n, etc.),
- Medicina (efectos de m√∫ltiples factores sobre una enfermedad),
- Ingenier√≠a (modelos de predicci√≥n de fallas o calidad),
- Ciencias sociales (relaci√≥n entre nivel educativo, ingresos, etc.).

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['Age','Exercise Level','Social Interaction Score']]
Y = df['Happiness Score']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()

Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```

```python
b0, b1, b2, b3 = modelo.params

Fun = lambda X1, X2, X3 : b0 + b1 * X1 + b2 * X2 + b3 * X3

Fun(10, 50, 150)
```

```python
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan(modelo.resid, X_train_constante)
print(f'Valor-p de Breusch-Pagan: {valor_p_bp: 0.4f}')
```

Al analizar el conjunto de datos sobre salud mental y h√°bitos de vida, se observ√≥ que las correlaciones entre variables como el ejercicio, la interacci√≥n social y la edad son muy bajas. Esto indica que dichas variables no est√°n fuertemente relacionadas entre s√≠ ni con el estado psicol√≥gico de las personas. En consecuencia, se interpreta que realizar ejercicio o tener una vida social m√°s activa no necesariamente guarda una relaci√≥n clara con el bienestar mental de los individuos.

En el modelo de regresi√≥n, el coeficiente de determinaci√≥n fue de apenas 0.0063, lo que significa que el modelo solo explica el 0.63% del comportamiento de la variable dependiente. Este resultado evidencia que el modelo pr√°cticamente no logra explicar la variabilidad de dicha variable, por lo que se considera conveniente probar con otro tipo de modelo o incorporar nuevas variables.
Finalmente, la prueba de Breusch-Pagan arroj√≥ un valor-p de 0.8083, por lo que no se rechaza la hip√≥tesis nula. Esto sugiere que no existen problemas de heterocedasticidad en el modelo.

Como conclusi√≥n, el an√°lisis nos muestra que las variables consideradas no se relacionan de forma significativa con el estado psicol√≥gico, y el modelo de regresi√≥n utilizado no logra explicar la variable.Lo que plantea la necesidad de usar otro modelo para ver si asi se logra explicar mejor  los factores que influyen en la salud mental.
